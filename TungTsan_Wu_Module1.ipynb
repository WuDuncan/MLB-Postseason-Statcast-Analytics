```python
# Drop deprecated / unused columns
drop_cols = [
    "spin_dir", "spin_rate_deprecated", "break_angle_deprecated",
    "break_length_deprecated", "tfs_deprecated", "tfs_zulu_deprecated",
    "umpire", "sv_id"
]
# Also drop any "fielder_2.X" pattern columns (equivalent to -matches())
fielder2_cols = [c for c in df_mlb.columns if c.startswith("fielder_2.")]
clean_df = df_mlb.drop(columns=drop_cols + fielder2_cols, errors="ignore")

# Parse date
clean_df["game_date"] = pd.to_datetime(clean_df["game_date"])

# Convert to categorical (equivalent to as.factor())
factor_cols = [
    "pitch_type", "pitch_name", "description", "events",
    "if_fielding_alignment", "of_fielding_alignment",
    "zone", "p_throws", "stand", "inning_topbot", "type", "bb_type"
]
for col in factor_cols:
    if col in clean_df.columns:
        clean_df[col] = clean_df[col].astype("category")

# Base runner indicators: NA → 0, any value → 1
for base in ["on_1b", "on_2b", "on_3b"]:
    clean_df[base] = clean_df[base].notna().astype(int)

# Fill NaN categories (equivalent to fct_explicit_na())
clean_df["events"]  = clean_df["events"].cat.add_categories("no_event").fillna("no_event")
clean_df["bb_type"] = clean_df["bb_type"].cat.add_categories("no_hit").fillna("no_hit")
```
